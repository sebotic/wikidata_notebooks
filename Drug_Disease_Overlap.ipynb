```python
import pandas as pd
```

```python
sawada = pd.read_csv('sawada.csv')
fu = pd.read_csv('fu.csv')
cheng = pd.read_csv('cheng.csv')
```

```python
print(sawada.head())
print(fu.head())
print(cheng.head())
```

```python
sawada_set = {(x['drug'], x['target']) for c, x in sawada.iterrows()}
```

```python
len(sawada_set)
```

```python
fu_set = {(x['drug'], x['target']) for c, x in fu.iterrows()}
cheng_set = {(x['drug'], x['target']) for c, x in cheng.iterrows()}

print(len(fu_set))
print(len(cheng_set))
```

```python
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
from matplotlib_venn import venn2, venn3

plt.figure(figsize=(40,40))

font = {'family' : 'Source Sans Pro',
        'weight' : 'bold',
        'size'   : 28}

matplotlib.rc('font', **font)


venn2([fu_set, sawada_set], ('Fu et al', 'Sawada et al'))

plt.show()
```

```python
plt.figure(figsize=(40,40))

font = {'family' : 'Source Sans Pro',
        'weight' : 'bold',
        'size'   : 28}

matplotlib.rc('font', **font)


venn2([cheng_set, sawada_set], ('Cheng et al', 'Sawada et al'))

plt.show()
```

```python
plt.figure(figsize=(40,40))

font = {'family' : 'Source Sans Pro',
        'weight' : 'bold',
        'size'   : 28}

matplotlib.rc('font', **font)


venn2([fu_set, cheng_set], ('Fu et al', 'Cheng et al'))

plt.show()
```

```python
plt.figure(figsize=(40,40))

font = {'family' : 'Source Sans Pro',
        'weight' : 'bold',
        'size'   : 28}

matplotlib.rc('font', **font)


venn3([fu_set, sawada_set, cheng_set], ('Fu et al', 'Sawada et al', 'Cheng et al'))

plt.show()
```

```python
fu_drugs = set(fu['drug'])
fu_targets = set(fu['target'])
```

```python
print(len(fu_drugs))
len((fu_targets))
```

```python
sawada_drugs = set(sawada['drug'])
sawada_targets = set(sawada['target'])
```

```python
print(len(sawada_drugs))
len((sawada_targets))
```

```python
cheng_drugs = set(cheng['drug'])
cheng_targets = set(cheng['target'])
```

```python
plt.figure(figsize=(40,40))

font = {'family' : 'Source Sans Pro',
        'weight' : 'bold',
        'size'   : 28}

matplotlib.rc('font', **font)


venn3([fu_drugs, sawada_drugs, cheng_drugs], ('Fu et al', 'Sawada et al', 'Cheng et al'))
plt.title("Drug set overlaps")

plt.show()
```

```python
plt.figure(figsize=(40,40))

font = {'family' : 'Source Sans Pro',
        'weight' : 'bold',
        'size'   : 28}

matplotlib.rc('font', **font)


venn3([fu_targets, sawada_targets, cheng_targets], ('Fu et al', 'Sawada et al', 'Cheng et al'))
plt.title("Target set overlaps")

plt.show()
```
